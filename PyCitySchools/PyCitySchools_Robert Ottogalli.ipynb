```python
# Import dependencies 
import pandas as pd

# Import CSVs
schools_csv = "Resources/schools_complete.csv"
students_csv = "Resources/students_complete.csv"

# Read CSVs as data frames
school_df = pd.read_csv(schools_csv)
students_df = pd.read_csv(students_csv)

# Merge school and student data frames into single data frame
py_schools_df = pd.merge(school_df, students_df, on="school_name")
py_schools_df.head()
```

```python
# Clean data
# Rename Columns
py_schools_df.columns
renamed = py_schools_df.rename(columns ={
                     'school_name':'School Name', 
                     'type': 'School Type', 
                     'size':'Total Students', 
                     'budget':'Total Budget', 
                     'student_name':'Student Name', 
                     'gender':'Gender', 
                     'grade':"Grade", 
                     'reading_score':'Reading Score', 
                     'math_score':"Math Score"})
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
### Calculate Data for District Summary

# Find total number of schools
Schools = renamed["School ID"].unique()
Total_Schools = len(Schools)

# Find total number of students
Total_Students = renamed["Student Name"].count()

# Find Total Budget
school_budget = renamed["Total Budget"].unique()
Total_Budget = sum(school_budget)

# Find Average Math Score
Average_Math = renamed["Math Score"].mean()

# Find Average Reading Score
Average_Reading = renamed["Reading Score"].sum()
```

```python
# Find % Passing Math 
passing_math = renamed.loc[renamed["Math Score"] >= 70].count(axis=1)  # Find number of passing reading scores as a dataframe
passing_math = passing_math.count()                                    # Convert data frame to count
Percent_Passing_Math = passing_math/Total_Students*100                 # Percent passing is count of passing scores, divided by total students, times 100

# Find % Passing Reading 
passing_reading = renamed.loc[renamed["Reading Score"] >= 70].count(axis=1)  # Find number of passing reading scores as a dataframe
passing_reading = passing_reading.count()                                    # Convert data frame to count
Percent_Passing_Reading = passing_reading/Total_Students*100                 # Percent passing is count of passing scores, divided by total students, times 100

# Find Overall Passing Rate (Average of % Passing Reading and % Passing Math)
Overall_Passing = (Percent_Passing_Math + Percent_Passing_Reading)/2
```

```python
# Create summary table for District Summary
# Create dictionary to hold summary data
Summary = {"Total Schools":[Total_Schools],
           "Total Students": [Total_Students],
           "Total Budget": [Total_Budget],
           "Average Math Score": [Average_Math],
           "Average Reading Score": [Average_Reading],
           "% Passing Math": [Percent_Passing_Math],
           "% Passing Reading": [Percent_Passing_Reading],
           "Overall Passing Rate": [Overall_Passing]}
# Convert summary dictionary to dataframe
Summary_Table = pd.DataFrame(Summary)

# Output Distric Summary Table
Summary_Table
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
### Calculate Data for School Summary

# Create a new variable to hold the existing data frame.
summary_df = renamed
summary_df

# Find per student budget
Per_Student_Budget = summary_df["Total Budget"]/summary_df["Total Students"]

# Add Per Student Budget to summary datafraome
summary_df.insert(5, "Per Student Budget", Per_Student_Budget, True)
```

```python
# # Create a data frame listing School Name and School Type
School_Summary = summary_df.groupby(["School Name", "School Type"])

# Adapt Total Students to fit grouped dataframe, select "mean" function
Total_Students = School_Summary["Total Students"].mean()

# Adapt Total School Budget to fit grouped dataframe, select "mean" function
Total_School_Budget = School_Summary["Total Budget"].mean()

# Adapt Per Student Budget to fit grouped dataframe, select "mean" function
Per_Student_Budget = School_Summary["Per Student Budget"].mean()

# Find average reading score for each school.  Use groupby School Name, select "mean" function
Average_Reading = School_Summary["Reading Score"].mean()

# Find average math score for each school.  Use groupby School Name, select "mean" function
Average_Math = School_Summary["Math Score"].mean()
```

```python
# Find % of Passing Reading grades per school

# Find Total Reading per school
Total_Scores = School_Summary.count()
Total_Scores

# Find passing reading scores per school
passing_reading = summary_df.loc[(summary_df["Reading Score"]>=70)]
passing_reading_count = passing_reading.groupby(["School Name", "School Type"]).count()

# Calculate Percent
Percent_Passing_Reading = (passing_reading_count/Total_Scores)*100
```

```python
# Find % of Passing Math grades per school

# Find passing math scores per school
passing_math = summary_df.loc[(summary_df["Math Score"]>=70)]
passing_math_count = passing_math.groupby(["School Name", "School Type"]).count()

# Calculate Percent
Percent_Passing_Math = (passing_math_count/Total_Scores)*100
```

```python
# Calculate overall passing rate (average of passing math and passing reading)
Overall_Passing = (Percent_Passing_Reading + Percent_Passing_Math)/2
```

```python
# Create dataframe to hold results for School Summary
School_Summary_Table = pd.DataFrame({
                            "Total Students":Total_Students,
                            "Total Budget": Total_School_Budget,
                            "Per Student Budget": Per_Student_Budget,
                            "Average Math Score": Average_Math,
                            "Average Reading Score": Average_Reading,
                            "% Passing Reading": Percent_Passing_Reading["Reading Score"],
                            "% Passing Math":  Percent_Passing_Math["Math Score"],
                            "Overall Passing Rate": Overall_Passing["Grade"]
                            })
```

```python
# Clean the formatting
School_Summary_Table_clean = School_Summary_Table.copy()
School_Summary_Table_clean["Total Budget"] = School_Summary_Table_clean["Total Budget"].map("${:,.2f}".format)
School_Summary_Table_clean["Per Student Budget"] = School_Summary_Table_clean["Per Student Budget"].map("${:.2f}".format)

# Output School Summary Table
School_Summary_Table_clean
```

## Math Scores by Grade
* Create a table that lists the average Math Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Set new dataframe to hold data from School Summary Table
Math_Table = summary_df
```

```python
# Create functions to find math grades of each student. Filter by grade, group by School name, hold as a Series.
# Create 1 function for each grade.  

def find_9th_math():                                                               # Function to find math scores of 9th graders
    ninth_grade = Math_Table.loc[(Math_Table["Grade"]=="9th"), ["Math Score"]]     # Find all Math Scores of 9th grade students
    ninth_grade = Math_Table.loc[(Math_Table["Grade"]=="9th"), :]                  # Select all rows identified in the previous row
    ninth_grade = ninth_grade.set_index("School Name")                             # Set Index to School Name
    ninth_grade = ninth_grade["Math Score"]                                        # Reduce number of columns to only hold index and Math Score column
    ninth_grade = ninth_grade.groupby("School Name")                               # Group by School Name
    return ninth_grade.mean()                                                      # Return Mean of math scores of 9th graders

def find_10th_math():
    tenth_grade = Math_Table.loc[(Math_Table["Grade"]=="10th"), ["Math Score"]]    # Repeat above proces, for 10th graders
    tenth_grade = Math_Table.loc[(Math_Table["Grade"]=="10th"), :]
    tenth_grade = tenth_grade.set_index("School Name")
    tenth_grade = tenth_grade["Math Score"]
    tenth_grade = tenth_grade.groupby("School Name")
    return tenth_grade.mean()

def find_11th_math():
    eleventh_grade = Math_Table.loc[(Math_Table["Grade"]=="11th"), ["Math Score"]] # Repeat above proces, for 11th graders
    eleventh_grade = Math_Table.loc[(Math_Table["Grade"]=="11th"), :]
    eleventh_grade = eleventh_grade.set_index("School Name")
    eleventh_grade = eleventh_grade["Math Score"]
    eleventh_grade = eleventh_grade.groupby("School Name")
    return eleventh_grade.mean()

def find_12th_math():
    twelfth_grade = Math_Table.loc[(Math_Table["Grade"]=="12th"), ["Math Score"]] # Repeat above proces, for 10th graders
    twelfth_grade = Math_Table.loc[(Math_Table["Grade"]=="12th"), :]
    twelfth_grade = twelfth_grade.set_index("School Name")
    twelfth_grade = twelfth_grade["Math Score"]
    twelfth_grade = twelfth_grade.groupby("School Name")
    return twelfth_grade.mean()


# Set variables to hold average math scores for each grade (defined in functions above)
ninth_math = find_9th_math()
tenth_math = find_10th_math()
eleventh_math = find_11th_math()
twelfth_math = find_12th_math()
```

```python
# Create dataframe to hold results for Math Scores by Grade
Math_Summary_Table = pd.DataFrame({
                            "9th":ninth_math,
                            "10th":tenth_math,
                            "11th":eleventh_math,
                            "12th":twelfth_math
                            })
# Output Math Scores by Grade
Math_Summary_Table
```
